```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
data["Cloud_Cover_encoded"]= label_encoder.fit_transform(data['Cloud Cover'])
data["Season_encoded"]= label_encoder.fit_transform(data['Season'])
data["Location_encoded"]= label_encoder.fit_transform(data['Location'])
data["Weather_Type_encoded"]= label_encoder.fit_transform(data['Weather Type'])
data
```

```python
data.head(50)
```

```python
#Encoded variables
#Cloud Cover ["clear", 'overcast', 'partly cloudy']=[0,1,2]
#Season ['Autumn','Spring','Summer',' Winter']=[0,1,2,3]
#Location['Coastal', 'inland', 'mountain] '=[0,1,2]
#Weather Type [Cloudy, Rainy, Snowy]= [0,1,2,3]
```

```python
data.reset_index(drop=True, inplace=True)
X=data.drop(columns=['Cloud Cover', 'Season', 'Location', "Weather Type",'index', 'Weather_Type_encoded'])
Y=data[['Weather_Type_encoded']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(10,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(4, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train, Y_train, validation_split=0.1, epochs=10)
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(10,)),
                          keras.layers.Dense(256, activation='relu'),
                          keras.layers.Dense(128, activation='relu'),
                          keras.layers.Dense(64, activation='relu'),
                          keras.layers.Dense(4, activation='sigmoid')
])

# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=20)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```
